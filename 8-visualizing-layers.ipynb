```python
import keras
```

```python
model = keras.models.load_model("data/mnist_train.h5")
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
weights, bias = model.layers[0].get_weights()
w,h,c,f = weights.shape

plt.figure(figsize=(20,10))
for i in range(f):
    w = np.squeeze(weights[:,:,:,i],axis=2)
    plt.subplot(6,6,i+1)
    plt.imshow(w,cmap="gray")
```
